```python
html=requests.get("https://pubgtracker.com/leaderboards").text
```

```python
from bs4 import BeautifulSoup as bs
```

```python
bsobj = bs(html,"html.parser")
bsobj
```

```python
type(html)
```

```python
name = bsobj.find_all("td")
players = []
for i in name:
    for j in i.find_all('a',{'data-tooltip':"notooltip"}):
        players.append(j.text.strip())
```
